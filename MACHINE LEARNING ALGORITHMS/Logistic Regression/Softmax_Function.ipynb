```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# Transforming the output column
encoder = LabelEncoder()
df['species'] = encoder.fit_transform(df['species'])
```

```python
df.head()
```

```python
# Choose only the required columns
df = df[['sepal_length','petal_length','species']]
```

```python
df.head()
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# multinomial here means softmax
clf = LogisticRegression(multi_class='multinomial')
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred))
```

```python
# Prediction
query = np.array([[3.4,2.7]])
# What predict_proba returns is the probability of each class
clf.predict_proba(query)
```

### INSIGHT:
- The first probability is the probability of flower(query) being setosa
- The second probability is the probability of flower(query) being versicolor
- The third probability is the probability of flower(query) being virginica

```python
# we'll get max value as predicted class
clf.predict(query)
```

```python
# That is the class with highest probability is first class
# That mean flower is most likely to be setosa
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x.values,y.values,clf,legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('Softmax on iris')

plt.show()
```
