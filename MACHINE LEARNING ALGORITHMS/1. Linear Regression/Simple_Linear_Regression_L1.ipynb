# SIMPLE LINEAR REGRESSION

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../DATASET/placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

```python
#  separating input and output columns
# x= input columns and y= output columns
x= df.iloc[:,0:1] 
y= df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
y_pred= lr.predict(x_test)
y_pred
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_test,lr.predict(x_test),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(In Lacks)')
plt.show()
```

```python
m=lr.coef_
m
```

```python
b=lr.intercept_
b
```

```python
# y =mx + b
# lets verify the equation with some values
# lets take first cgpa (8.58) as input and see if it matches with the model output
y = m * 8.58 + b
print(y)
# the output matches the trained model output
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
