```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('../../DATASET/train.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

## Descision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tre = DecisionTreeClassifier()
```

```python
tre.fit(X_train,y_train)
```

```python
y_pred2 = tre.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

## RECALL , PRECISION , F1 SCORE

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

### Calculate Total precision,recall and f1 score

```python
# Total Precision
precision_score(y_test,y_pred1,average=None)
```

```python
# Toal Recall
recall_score(y_test,y_pred1,average=None)
```

```python
# Total F1 Score
f1_score(y_test,y_pred1,average=None)
```

### To calculate combined value of precision ,recall and f1 score
- 1. Macro
- 2. Weighted

#### MACRO:

```python
# Macro accuracy
precision_score(y_test,y_pred1,average='macro')
```

```python
# Macro recall
recall_score(y_test,y_pred1,average='macro')
```

```python
# Macro f1 score
f1_score(y_test,y_pred1,average='macro')
```

#### WEIGHTED:

```python
# Weighted accuracy
precision_score(y_test,y_pred1,average='weighted')
```

```python
# Weighted recall
recall_score(y_test,y_pred1,average='weighted')
```

```python
# Weighted f1 score
f1_score(y_test,y_pred1,average='weighted')
```

## To Calculate all the metrics at once

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
