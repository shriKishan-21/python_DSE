```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('pima-indians-diabetes.data.csv')
```

```python
df=pd.read_csv('pima-indians-diabetes.data.csv',index_col=0)
```

```python
df
```

```python
print(df.columns)
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df.drop('1',axis=1))
```

```python
scaled_features=scaler.transform(df.drop('1',axis=1))
```

```python
scaled_data=pd.DataFrame(scaled_features,columns=df.drop('1',axis=1).columns)
```

```python
# Split the data set into training and test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=scaled_data
y=df['1']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# Train the model and make prediction
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
predictions=model.predict(x_test)
```
